#### Task-6 Prediction using Decision Tree Algorithm

<h4 style="color:green"> <i>Importing libraries</i></h4>

```python
#importing all libraries required for this task
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
import seaborn as sn
from sklearn import tree
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix 
```

<h4 style="color:green"> <i>Reading Data</i></h4>

```python
#reading the iris dataset
data_set=pd.read_csv("Iris.csv")
print("The data has been imported, as shown below")
data_set.head()
```

<h3 style="color:green"> <i>Preparing the Data</i></h3>

<h4 style="color:green"> <i>Encoding the Categorical Data</i></h4>

```python
#Encoding with LabelEncoder
data_set['spe_enc']=LabelEncoder().fit_transform(data_set['Species'])
#dividing data into inputs and Outputs 
#that is features and target
X=data_set.iloc[:,1:5].values
y=data_set.iloc[:,-1].values
```

<h4 style="color:green"> <i>Feature Data</i></h4>

```python
print(X)
```

<h4 style="color:green"> <i>Target Data</i></h4>

```python
print(y)
```

<h4 style="color:green"> <i>Splitting the data into Training and Testing in the ratio 4 to 1 </i></h4>

```python
#splitting into X_train,X_test,y_train,y_test

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=1/4)
```

## Building Model

<h2 style="color:green"><i>Defining snd Fittitng the Decision Tree Algorithm</i></h2>

```python
dtcc=DecisionTreeClassifier()
dtcc.fit(X_train,y_train)
```

<h2 style="color:green"><i>Visualizing the Decision Tree Classifier on Training Datset </i></h2>


```python
plt.figure(figsize=(25,20),facecolor="pink")
tree.plot_tree(dtcc,
              feature_names=list(data_set.iloc[:,1:5].columns),
              class_names=list(data_set.Species.unique()),
               filled=True)

tree
```

<h4 style="color:green"> <i>Making Predictions on Testing Data</i></h4>


```python
#making predictions using predict()
y_predict=dtcc.predict(X_test)
print(y_predict)
```

### Checking the Accuracy by  Confusion Matrix

```python
#confusion matrix compares the Actual and Predicted Data
cm= confusion_matrix(y_test, y_predict)
print(cm)
plt.figure(figsize=(10,7),facecolor="lightblue")
sn.heatmap(cm,annot=True)
plt.ylabel("Actual Data",fontsize="20")
plt.xlabel("Predicted Data",fontsize="20")
plt.title("Confusion Matrix(Iris dataset)",fontsize="30")
```

```python
accuracy=np.trace(cm)/np.sum(cm)
print(accuracy*100)
```
